```python
# prompt: crear 3 variables en cada una guarda valores posteriormente sacamos el valor minimo el maximo y el valor promedio

# Crear 3 variables con valores
variable1 = [10, 20, 30]
variable2 = (5, 15, 25)
variable3 = {1: 5, 2: 10, 3: 15}

# Encontrar el valor mÃ­nimo, mÃ¡ximo y promedio de cada variable
min_v1 = min(variable1)
max_v1 = max(variable1)
prom_v1 = sum(variable1) / len(variable1)

min_v2 = min(variable2)
max_v2 = max(variable2)
prom_v2 = sum(variable2) / len(variable2)

min_v3 = min(variable3.values())
max_v3 = max(variable3.values())
prom_v3 = sum(variable3.values()) / len(variable3)

# Imprimir los resultados
print("Variable 1: Minimo =", min_v1, ", Maximo =", max_v1, ", Promedio =", prom_v1)
print("Variable 2: Minimo =", min_v2, ", Maximo =", max_v2, ", Promedio =", prom_v2)
print("Variable 3: Minimo =", min_v3, ", Maximo =", max_v3, ", Promedio =", prom_v3)
```
